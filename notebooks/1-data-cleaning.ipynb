## Czyszczenie danych

```python
df.drop_duplicates(inplace=True)
```

```python
df = df[~df['year'].isna()].copy()
df = df[(2000 <= df['year']) & (df['year'] < 2020)].copy()
df = df[1000 < df['price']].copy()
df = df[0 <= df['odometer']].copy()
df = df[df['title_status'] == 'clean'].copy()
df = df[df['condition'] != 'salvage'].copy()
df = df[~df['model'].isna()].copy()
df = df[~df['posting_date'].isna()].copy()
```

```python
df['posting_date'] = pd.to_datetime(df['posting_date'], utc=True)
df['posting_date'].max(), df['posting_date'].min()
```

```python
to_drop = [
    'VIN',
    'id',
    'image_url',
    'region_url',
    'url',
    'posting_date',
    'state',
    'county',
    'size',
    'description',
    'region',
    'lat',
    'long',
    'title_status',
    'model'
]

df.drop(to_drop, axis=1, inplace=True)
```

```python
def remove_outliers(df, col):
    k = 1.5
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)    
    iqr = q3 - q1
    df = df[(q1 - k * iqr < df[col]) & (df[col] < q3 + k * iqr)]
    return df

df = remove_outliers(df, 'price')
df = remove_outliers(df, 'odometer')
```

```python
def fill_na(df, col):
    if pd.api.types.is_string_dtype(df[col].dtype):
        df.loc[:, col] = df[col].fillna(df[col].mode().to_list()[0])
    else:
        try:
            df.loc[:, col] = df[col].fillna(df[col].median())
        except Exception as e:
            df.loc[:, col] = df[col].fillna(df[col].median().to_list()[0])
    return df

missing = pd.DataFrame(
    data=df.isna().sum() / len(df) * 100, 
    columns=['Pct']
)
missing.sort_values('Pct', ascending=False, inplace=True)

for col in missing[missing['Pct'] > 0].index:
    before = df[col].isna().sum()
    df = fill_na(df, col)
    after = df[col].isna().sum()    
    print(f'{col:<20} {before:<10} {after:<10} {before - after:<10}') 
```

```python
print(df.shape)
manu = df[['manufacturer', 'price']].groupby('manufacturer', as_index=False).size()
manu.sort_values('size', inplace=True, ascending=False)

to_small_cat = manu[manu['size'] < manu['size'].quantile(0.25)]['manufacturer'].to_list()
df = df[~df['manufacturer'].isin(to_small_cat)]  # Usuwamy rzadkich producentów

print(to_small_cat)
print(df.shape)
```

```python
df.drop_duplicates(inplace=True)
```

## Wykresy słupkowe

```python
target = 'price'
for col in  df.select_dtypes('object'):
    if col != 'price':
        grouped = df[['price', col]].dropna().groupby(col, as_index=False).mean()
        grouped.sort_values('price', ascending=False, inplace=True)
        plt.figure(figsize=(4 * (16/9),4), dpi=150)
        plt.bar(grouped[col], grouped['price'])
        plt.xticks(rotation=90)
        plt.title(f'{col.capitalize()} vs Mean price')
        plt.tight_layout()
        plt.savefig(f'../reports/figures/clean/{col}_vs_mean_price.png')
```
